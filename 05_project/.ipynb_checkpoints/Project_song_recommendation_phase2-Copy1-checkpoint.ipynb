### kmeans features Kaggle from Spotify

```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
from matplotlib import pyplot
```

```python
# read the csv and create the dataframe
path = 'data/features_kaggle.csv'
df = pd.read_csv(path)
```

```python
df.head()
```

```python
df.columns
```

```python
# Removes temporarily non numerical fetures. It is likely that I have to assign them at the end.
df = df[['energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo','duration_ms',
       'time_signature','danceability']].copy()
df.head(50)
```

```python
plt.scatter(df['energy'],df['danceability'])
plt.xlabel('energy')
plt.ylabel('danceability')
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

# # taking all rows but only 6 columns
# df_small = df.iloc[:,:6]

correlation_mat = df.corr()

sns.heatmap(correlation_mat, annot = True)

plt.show()
```

```python
plt.scatter(df['valence'],df['danceability'])
plt.xlabel('valence')
plt.ylabel('danceability')
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['valence','danceability']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.valence,df1['danceability'],color='green')
plt.scatter(df2.valence,df2['danceability'],color='red')
plt.scatter(df3.valence,df3['danceability'],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('valence')
plt.ylabel('danceability')
plt.legend()
```

```python
scaler = MinMaxScaler()

scaler.fit(df[['danceability']])
df['danceability'] = scaler.transform(df[['danceability']])

scaler.fit(df[['valence']])
df['valence'] = scaler.transform(df[['valence']])
```

```python
df.head()
```

```python
plt.scatter(df.valence,df['danceability'])
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['valence','danceability']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.valence,df1['danceability'],color='green')
plt.scatter(df2.valence,df2['danceability'],color='red')
plt.scatter(df3.valence,df3['danceability'],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.legend()
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['valence','danceability']])
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
