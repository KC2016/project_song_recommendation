### Create a function that check if the song belong to the top 100 song 
Based on scraped info from the url = 'https://www.billboard.com/charts/hot-100'

I create a function that 
- input a song 
- check if the user typed something and

- if the song is in the list it print a message or
- if is not in the list, it gives a random recommendation of one of the top 100 songs.</br>



```python
import pandas as pd
from bs4 import BeautifulSoup 
import requests
from random import randint
```

```python
path = 'data/Billboard_scraping.csv'
top100_df = pd.read_csv(path)
top100_df
```

```python
# define a function

def recommender():
    # input a song
    song_choice = input("what is the song? ")
    # answer if the song is in the list of the top 100
    if len(song_choice)==0:
        print("Please, write a song: ")
        
    else: 
        if song_choice in list(top100_df['Song']):
            print('You have a good taste, this song is hot')
        else:
            print('Wrong choice, this song is not hot')
            random_song = randint(0, len(top100_df)-1)
            print("You could also like " + top100_df["Song"][random_song] + " by " + top100_df["Artists"][random_song])
    
recommender()
```
